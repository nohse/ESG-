```python
def esg_2023_scraper():
    driver = webdriver.Chrome()
    driver.get('https://esg.krx.co.kr/contents/02/02020000/ESG02020000.jsp')
    time.sleep(10) # 웹사이트가 켜지는 시간동안 잠시 멈춤 합니다.

    global data_df_2023 # ****** 부분에 전역변수 설정 코드를 작성합니다.
    data_df_2023 = pd.DataFrame(columns=("회사명", "KCGS_종합등급", "KCGS_환경", "KCGS_사회", "KCGS_지배구조", "한국ESG연구소", "Moody's", "MSCI", "S&P", "연도"))

    index = 0

    while index < 794: # 총 페이지가 794장 있기 떄문에 <794로 제한을 둡니다.
        # 각 페이지의 테이블 가져오기
        table = driver.find_element(By.CSS_SELECTOR,'#gridtableeccbc87e4b5ce2fe28308fd9f2a7baf3')
        tbody = table.find_element(By.CSS_SELECTOR,"tbody")
        rows = tbody.find_elements(By.CSS_SELECTOR,"tr")
        # 한 페이지 속 기업들(rows)에 대해
        for i in range(len(rows)):
            # 매번 새로운 rows를 가져옵니다.
            rows = tbody.find_elements(By.CSS_SELECTOR,"tr") # 위에 rows = ******.find_elements(******) 랑 동일
            row = rows[i]
            # 기업명
            company_name = row.find_element(By.CSS_SELECTOR, "td:nth-child(1) > a").text
            print(company_name)
            # KCGS_종합 등급
            kcgs_1 = row.find_element(By.CSS_SELECTOR, "td:nth-child(2) > span").text
            print(kcgs_1)
            # KCGS_환경 등급
            kcgs_2 = row.find_element(By.CSS_SELECTOR, "td:nth-child(3) > span").text
            print(kcgs_2)
            # KCGS_사회 등급
            kcgs_3 = row.find_element(By.CSS_SELECTOR, "td:nth-child(4) > span").text
            print(kcgs_3)
            # KCGS_지배구조 등급
            kcgs_4 = row.find_element(By.CSS_SELECTOR, "td:nth-child(5) > span").text
            print(kcgs_4)
            #한국ESG연구소 등급
            kesg = row.find_element(By.CSS_SELECTOR, "td:nth-child(6) > span").text
            print(kesg)
            # Moody's 등급
            moodys = row.find_element(By.CSS_SELECTOR, "td:nth-child(8) > span").text
            print(moodys)
            # MSCI 등급
            msci = row.find_element(By.CSS_SELECTOR, "td:nth-child(9) > span").text
            print(msci)

            sp = row.find_element(By.CSS_SELECTOR, "td:nth-child(10) > span").text
            print(sp)
            # 연도
            year = 2023
            print(year)
            # 기업별 추출한 정보를 dataframe의 한 행으로 저장
            data_df_2023.loc[index] = [company_name, kcgs_1, kcgs_2, kcgs_3, kcgs_4, kesg, moodys, msci, sp, year]
            print("--------------------------------------")
            # index를 증가시키기 -> while이 반복되며 index가 794가 되면 종료
            index += 1

        # 한 페이지 내에는 10개의 row(기업) 존재,
        # 10개 기업 데이터 모두 추출한 후에는 다음 페이지로 이동
        if index % 10 == 0:
            next_page = driver.find_element(By.CSS_SELECTOR, "#pagenavia87ff679a2f3e71d9181a67b7542122c > ul > li.next > a")
            driver.execute_script("arguments[0].click();", next_page)
            time.sleep(10)

    print(data_df_2023)
    return data_df_2023
```
